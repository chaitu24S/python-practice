```python
sales = [
    {"city":"Mumbai","amount":1000},
    {"city":"Delhi","amount":800},
    {"city":"Mumbai","amount":500},
    {"city":"Pune","amount":700}
]
```

```python
sales_by_city = {}

for dtl in sales:
    city = dtl['city']
    amt = dtl['amount']

    if city in sales_by_city:
        sales_by_city[city]['amount'] += amt
        sales_by_city[city]['count'] += 1
    else:
        sales_by_city[city] = {
            'amount' : amt,
            'count' : 1
        }
```

```python
print(sales_by_city)
```

```python
for cty_nme, cty_detail in sales_by_city.items():
    avg = cty_detail['amount'] / cty_detail['count']
    sales_by_city[cty_nme]['avg_sale'] = avg
```

```python
print(sales_by_city)
```

```python
max_cty_sls = max(sales_by_city, key=lambda name: sales_by_city[name]['amount'])
cty_name, cty_amt = max_cty_sls, sales_by_city[max_cty_sls]['amount']
print("Highest revenue city:", cty_name,"revenue is:", cty_amt)
```

```python
for cty_nme, cty_dtl in sales_by_city.items():
    print("City name:", cty_nme, "with average revenue", cty_dtl['avg_sale'])
```

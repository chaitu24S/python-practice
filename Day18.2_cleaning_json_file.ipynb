```python
import json
```

```python
with open("employee_details.json", "r") as file:
    data = json.load(file)
```

```python
print(data)
```

```python
for row in data:
    print(row)
```

```python
cleaned_data = []

for row in data:

    # Validate age
    try:
        age = int(row['age'])
    except (ValueError, TypeError):
        continue

    if (
        row['id'] == 0 or
        row['name'] == '' or
        row['email'] == 'test@test.com' or
        age >= 50 or
        row['salary'] == 0 or
        row['department'] == ''
    ):
        continue

    row['age'] = age
    cleaned_data.append(row)

print(cleaned_data)
```

```python
sales = [
    {"customer":"Alice","amount":500},
    {"customer":"Bob","amount":700},
    {"customer":"Alice","amount":300},
    {"customer":"Charlie","amount":1000}
]
```

```python
sales_by_person = {}

for cust in sales:
    customer = cust["customer"]
    amount = cust["amount"]

    if customer in sales_by_person:
        sales_by_person[customer]['total'] += amount
        sales_by_person[customer]["count"] += 1
    else:
        sales_by_person[customer] = {
            'total' : amount,
            'count' : 1
        }

print(sales_by_person)
```

```python
total_amt = []
for cust in sales_by_person.values():
    total_amt.append(cust['total'])

print("Total sales amount:", sum(total_amt))
```

```python
for cust_name, cust_detail in sales_by_person.items():
    average = cust_detail['total'] / cust_detail['count']
    sales_by_person[cust_name]['avg'] = average
```

```python
print(sales_by_person)
```

```python
top_cust_name = max(sales_by_person, key=lambda name:sales_by_person[name]['total'])
cust_name, cust_amt = top_cust_name, sales_by_person[top_cust_name]['total']
print("Max total salesman:", cust_name, "amount:", cust_amt)
    
```

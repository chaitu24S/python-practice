```python
from pyspark.sql import SparkSession
import pandas as pd
```

```python
spark = SparkSession.builder.appName("Practice").getOrCreate()

print(spark.version)
```

```python
data = [
    (1, "John", 50000),
    (2, "Alice", 60000),
    (3, "Bob", 45000)
]
```

```python
df = spark.createDataFrame(data, ["emp_id", "name", "salary"])

df.show()
```

```python
df.show(2)
```

```python
df.printSchema()
```

```python
df.select('name', 'salary').show()
```

```python
df.select("name", (df.salary + 5000).alias("new_salary")).show()
```

```python
df.filter(df.salary > 50000).show()
```

```python
df.filter("salary > 50000").show()
```

```python
df.filter((df.salary > 45000) & (df.name != "John")).show()
```

```python
import json
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, avg
```

```python
spark = SparkSession.builder.appName("Practice").getOrCreate()

print(spark.version)
```

```python
df = spark.read.option("multiline", "true").json("employee_details.json")

df.show()
```

```python
df.printSchema()
```

```python
df = df.withColumn("salary", col("salary").cast('double'))
```

```python
df.printSchema()
```

```python
valid_df = df.filter("salary > 0")
```

```python
valid_df.show()
```

```python
df.show()
```

```python
df.select("age").show()
```

```python
df = df.withColumn("age", when(col("age") == "Twenty Five", "25").otherwise(col("age")))

df.show()
```

```python
df = df.withColumn("age", when(col("age") == "", None).otherwise(col("age")))

df.show()
```

```python
df = df.withColumn("age", col("age").cast("int"))

df.show()
```

```python
df= df.filter(col("age").isNull() | (col("age") <= 100))

df.show()
```

```python
avg_age = df.select(avg("age")).first()[0]

print(avg_age)
```

```python
df = df.na.fill({"age": int(avg_age)})

df.show()
```

```python
transactions = [
    {"customer":"Alice","amount":500},
    {"customer":"Bob","amount":700},
    {"customer":"Alice","amount":300},
    {"customer":"Charlie","amount":1000},
    {"customer":"Bob","amount":200}
]
```

```python
total_sales = []
for amt in transactions:
    amount = amt['amount']
    total_sales.append(amount)

print(total_sales)
```

```python
print("Total sales:", sum(total_sales))
```

```python
total_sales_by_customer = {}

for cust in transactions:
    customer = cust["customer"]
    amount = cust["amount"]

    if customer in total_sales_by_customer:
        total_sales_by_customer[customer] += amount
    else:
        total_sales_by_customer[customer] = amount

print(total_sales_by_customer)
```

```python
print("Average Transaction Amount:", sum(total_sales) / len(total_sales))
```

```python
for cust_600 in total_sales_by_customer.items():
    if cust_600[1] > 900:
        print("Customer total sales greater than 900 is", cust_600[0], "amount:", cust_600[1])
```

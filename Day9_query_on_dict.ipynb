```python
orders = [
    {"customer":"Alice","amount":500},
    {"customer":"Bob","amount":700},
    {"customer":"Alice","amount":300},
    {"customer":"Charlie","amount":1000},
    {"customer":"Bob","amount":200}
]
```

```python
sales = {}

for cust in orders:
    customer = cust["customer"]
    amount = cust["amount"]
   
    if customer in sales:
        sales[customer] += amount
    else:
        sales[customer] = amount

print(sales)
```

```python
customer, amount = max(sales.items(), key=lambda item:item[1])
print("Max customer name:", customer , "max sales:", amount)
```

```python
sales = {}

for cust in orders:
    customer = cust["customer"]
    amount = cust["amount"]

    if customer in sales:
        sales[customer]['total'] += amount
        sales[customer]["count"] += 1
    else:
        sales[customer] = {
            "total": amount,
            "count": 1
        }

print(sales)
```

```python
for customer, info in sales.items():
    avg = info["total"] / info["count"]
    print(customer, avg)
```

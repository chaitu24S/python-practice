```python
import json
```

```python
with open("employee.json", "r") as files:
    data = json.load(files)
```

```python
print(data)
```

```python
dpt_lst = {}

for dept in data:
    sector = dept['dept']
    ttl_slry = dept['salary']

    if sector in dpt_lst:
        dpt_lst[sector]['total_slry'] += ttl_slry
        dpt_lst[sector]['count'] += 1
    else:
        dpt_lst[sector]= {
            'total_slry' : ttl_slry,
            'count' : 1
        }
```

```python
print(dpt_lst)
```

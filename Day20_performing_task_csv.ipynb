```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, sum, avg, desc
```

```python
spark  =  SparkSession.builder.appName("Practice").getOrCreate()

print(spark.version)
```

```python
df = spark.read.csv("sales_data.csv", header=True)

df.show()
```

```python
df.printSchema()
```

```python
df.select(col("quantity"), col("price"), col("discount")).show()
```

```python
targeted_colmn = ['quantity', 'price','discount']
df.select(*targeted_colmn).show()
```

```python
df.select('quantity', 'price', 'discount').show()
```

```python
df = df.withColumnRenamed('payment_method', 'payment_mode')

df.show()
```

```python
df.filter('quantity > 2').show()
```

```python
df.filter(df.city == 'Mumbai').show()
```

```python
df.printSchema()
```

```python
df = df.withColumn('quantity', col('quantity').cast('int'))
df = df.withColumn('price', col('price').cast('int'))
df = df.withColumn('discount', col('discount').cast('int'))

df.printSchema()
```

```python
df.show()
```

```python
df = df.withColumn('final_price', col('price') - col('discount'))

df.show()
```

```python
category = df.groupBy('category').agg(sum("final_price").alias('grp_by_category'))
```

```python
category.show()
```

```python
product = df.groupBy("category").agg(count('product').alias('pdt_count')).show()
```

```python
avg_prc = df.groupBy("category").agg(avg('final_price').alias("avg_prc_prdct")).show()
```

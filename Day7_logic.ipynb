```python
orders = [
    {"customer":"Alice","amount":500},
    {"customer":"Bob","amount":700},
    {"customer":"Alice","amount":300},
    {"customer":"Charlie","amount":1000}
]
```

```python
total_sales = 0

for amt in orders:
    total_sales += amt["amount"]


print("Total sales:", total_sales)
```

```python
sales = {}

for cust in orders:
    customer = cust["customer"]
    amount = cust["amount"]

    if customer in sales:
        sales[customer] += amount
    else:
        sales[customer] = amount
print(sales)
```

```python
max_key, max_value = max(sales.items(), key=lambda item: item[1])

print("Customer name is:", max_key," order of:", max_value)
```

```python
sales_amount = list(sales.values())
average_sales_amount = sum(sales_amount) / len(sales_amount)
print(average_sales_amount)
```

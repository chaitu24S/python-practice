```python
# import the json module
import json 

# Read JSON file
with open("employee_revision.json", "r") as file:
    data = json.load(file)

print(data)

```

```python
# Sorting into list
employees = []
salaries = []
employees_detail = [detail for detail in data]
for employee in data:
    employees.append(employee["name"])
    salaries.append(employee["salary"])

print(employees_detail)
print(employees)
print(salaries)
```

```python
# Highest salary
print("Highest salary:", max(salaries))

# Lowest salary
print("Lowest salary:", min(salaries))

# Average salary
print("Average salary:", int(sum(salaries) / len(salaries)))
```

```python
highest_salary = max(data, key=lambda employee : employee["salary"])
lowest_salary = min(data, key=lambda employee : employee["salary"])
average_salary = sum(employee["salary"] for employee in data) / len(data)
```

```python
print("Highest salary:", highest_salary["name"], "-", highest_salary["salary"])
print("Lowest salary:", lowest_salary["name"], "-", lowest_salary["salary"])
print("Average salary:", int(average_salary))
```

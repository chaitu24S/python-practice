```python
lst = []

for i in range(5):
    num = int(input(f"Enter number {i +1}:"))
    lst.append(num)

print(lst)

n = 0
for i in lst:
    n += i

print(n)

print(len(lst))

def avg(x):
    total = sum(x)
    n_item = len(x)
    return total /n_item

avg(lst)

print(max(lst))
print(min(lst))
```

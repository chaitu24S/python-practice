```python
import json
```

```python
with open("employee.json", "r") as file:
    data = json.load(file)

print(data)
```

```python
dept_salaries = {}

for emp in data:
    dept = emp["dept"]
    salary = emp["salary"]

    if dept not in dept_salaries:
        dept_salaries[dept] = []

    dept_salaries[dept].append(salary)

for dept, salaries in dept_salaries.items():
    avg_salary = sum(salaries) / len(salaries)
    print(dept, ":", avg_salary)
```

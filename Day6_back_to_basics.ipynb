```python
numbers = [10, 20, 30, 40, 50]

print("Sum:", sum(numbers))
print("Maximum:", max(numbers))
print("Minimum:", min(numbers))
print("Average:", sum(numbers) / len(numbers))
```

```python
employees = [
    {"name":"Alice","salary":50000},
    {"name":"Bob","salary":70000},
    {"name":"Charlie","salary":60000}
]
```

```python
emp_salaries = []
max_salary_employee = max(employees, key = lambda employee : employee["salary"])

for emp in employees:
    emp_salaries.append(emp["salary"])

print(int(sum(emp_salaries) / len(emp_salaries)))
print("Name:", max_salary_employee["name"], "salary:", max_salary_employee["salary"])
```

```python
import csv
```

```python
with open("employees.csv", "r") as file:
    data = list(csv.reader(file))
```

```python
headers = data[0]

employees = []

for row in data[1:]:
    employee = dict(zip(headers, row))
    employees.append(employee)

print(employees)
```

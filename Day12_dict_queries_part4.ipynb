```python
transactions = [
    {"customer":"Alice","amount":500},
    {"customer":"Bob","amount":700},
    {"customer":"Alice","amount":300},
    {"customer":"Charlie","amount":1000},
    {"customer":"Bob","amount":200}
]
```

```python
sales_transactions = {}

for cust in transactions:
    customer = cust['customer']
    amt = cust['amount']

    if customer in sales_transactions:
        sales_transactions[customer]['amt'] += amt
        sales_transactions[customer]['count'] += 1
    else:
        sales_transactions[customer] = {
            'amt' : amt,
            'count' : 1
        }
```

```python
print(sales_transactions)
```

```python
for cust_name, cust_detail in sales_transactions.items():
    average = cust_detail['amt'] / cust_detail['count']
    sales_transactions[cust_name]['avg'] = average
```

```python
print(sales_transactions)
```

```python
highest_spending_cust = max(sales_transactions, key=lambda name : sales_transactions[name]['amt'])
print(highest_spending_cust)
cust_name, cust_total = highest_spending_cust, sales_transactions[highest_spending_cust]['amt']
print("Customer name with highest spending:", cust_name, "spending off:", cust_total)
```

```python
orders = [
    {"customer":"Alice","city":"Mumbai","amount":70000},
    {"customer":"Bob","city":"Delhi","amount":40000},
    {"customer":"Alice","city":"Mumbai","amount":10000},
    {"customer":"Charlie","city":"Pune","amount":5000},
    {"customer":"David","city":"Mumbai","amount":70000},
    {"customer":"Bob","city":"Delhi","amount":6000}
]
```

```python
ord_cst = {}

for cust in orders:
    customer = cust['customer']
    amt = cust['amount']
    cty = cust['city']

    if cty in ord_cst:
        ord_cst[cty]['amount'] += amt
        ord_cst[cty]['count'] += 1
    else:
        ord_cst[cty] = {
            'amount' : amt,
            'count' : 1
        }
```

```python
print(ord_cst)
```

```python
for cty_nme, cty_dtl in ord_cst.items():
    avg = cty_dtl["amount"] / cty_dtl["count"]
    ord_cst[cty_nme]["avg"] = int(avg)
```

```python
print(ord_cst)
```

```python
max_ord_cty = max(ord_cst, key=lambda name: ord_cst[name]['amount'])
max_city_nme, max_city_vlu = max_ord_cty, ord_cst[max_ord_cty]["amount"]
print("City with highest revenue:", max_city_nme, "total order of:", max_city_vlu)
```

```python
min_ord_cty = min(ord_cst, key=lambda name: ord_cst[name]['amount'])
min_city_nme, min_city_vlu = min_ord_cty, ord_cst[min_ord_cty]['amount']
print("City with lowest revenue:", min_city_nme, "total order of:", min_city_vlu)
```

```python
ovrl_sls = 0

for cty_nme, cty_dtl in ord_cst.items():
    ovrl_sls += cty_dtl['amount']
    
print("Overall sales is:", ovrl_sls)
```

```python
ovrl_sls = 0
ovrl_cnt = 0

for cty_nme, cty_dtl in ord_cst.items():
    ovrl_sls += cty_dtl['amount']
    ovrl_cnt += cty_dtl['count']
    
print("Overall sales average is:", ovrl_sls/ovrl_cnt)
```

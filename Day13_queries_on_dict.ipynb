```python
employees = [
    {"name":"Alice","dept":"IT","salary":50000},
    {"name":"Bob","dept":"IT","salary":70000},
    {"name":"Charlie","dept":"HR","salary":60000},
    {"name":"David","dept":"IT","salary":80000}
]
```

```python
dpt = {}

for empy in employees:
    sec = empy['dept']
    slry = empy['salary']

    if sec in dpt:
        dpt[sec]['count'] += 1
        dpt[sec]['total_salary'] += slry
    else:
        dpt[sec]= {
            'count' : 1,
            'total_salary' : slry
        }
```

```python
print(dpt)
```

```python
for dpt_name, dpt_detail in dpt.items():
    avg = dpt_detail["total_salary"] / dpt_detail["count"]
    dpt[dpt_name]['average_salary'] = avg

print(dpt)
```
